```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def get_all_marques():
    url='https://www.automobile.tn/fr/neuf'
    source=requests.get(url).text
    soup = BeautifulSoup(source, 'lxml')
    links=soup.findAll('a')
    marques=[]
    for link in links:
        if(len(link.findChildren("picture"))):
            marques.append(link['href'])
    return marques
```

```python
def get_data():    
    marques=get_all_marques()
    dictionnary={}
    df=pd.DataFrame(columns=['marque','car names','prices'])
    for marque in marques:
        lst_car_names=[]
        lst_prices=[]
        url='https://www.automobile.tn'+marque
        source=requests.get(url).text
        soup = BeautifulSoup(source, 'lxml')
        car_names=soup.findAll('h2')
        prices=soup.findAll('div',{"class": "price"})
        for price in prices:
            lst_prices.append(price.findChild('span').text)
        for car_name in car_names:
            lst_car_names.append(car_name.text)
        dictionnary={'marque':marque[9:],'car names':lst_car_names,'prices':lst_prices}
        temp=pd.DataFrame(dictionnary)
        df=df.append(temp)
    return df
```
