```python
import pandas as pd
import datetime
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys 
```

```python
def real_time_price(stock_code):
    url = ('https://finance.yahoo.com/quote/' + stock_code + '.HK?p=' + stock_code + '.HK&.tsrc=fin-srch')
    r=requests.get(url)
    web_content = BeautifulSoup(r.text, 'lxml')
    web_content = web_content.find('div', class_='My(6px) Pos(r) smartphone_Mt(6px)')
    web_content = web_content.find('span').text
    return web_content
```

```python
#Stock's code list
code_list = ['0001','0002', '0003', '0004']
#time to sleep
seconds=5
for step in range(1,5):
    #list for the prices
    price=[]
    #col for the date
    col=[]
    #getting the current date 
    time_stamp = datetime.datetime.now()
    #formatting the date so it's clearer
    time_stamp = time_stamp.strftime("%Y-%m-%d %H:%M:%S")
    #calling the function for each code
    for stock_code in code_list:
        price.append(real_time_price(stock_code))
    #appending the prices with their respective date
    col = [time_stamp]
    col.extend(price)
    #coverting the data to a dataframe
    df = pd.DataFrame(col)
    #transposing the dataframe
    df = df.T
    print(col)
    #seconds minus 6 since the scrapping take on average 6-7 seconds
    if (seconds>6):
        seconds=seconds-6
    #if the seconds are less than 6 we can't substitute or we will have a negative argument
    else:
        seconds=0
    time.sleep(seconds)
```

```python
#getting historical data 
chromedriver_path= "C:/Users/ASUS/Downloads/chromedriver_win32/chromedriver"
driver = webdriver.Chrome(chromedriver_path)
#url = ('https://finance.yahoo.com/quote/' + stock_code + '.HK?p=' + stock_code + '.HK&.tsrc=fin-srch')
url = ('https://finance.yahoo.com/quote/0001.HK/history?p=0001.HK')
driver.get(url)
driver.find_element_by_xpath("//div[@class='M(0) O(n):f D(ib) Bd(0) dateRangeBtn O(n):f']").click()
driver.find_element_by_xpath("//button[@data-value='MAX']").click()
data = []
for step in range(1,10):
    driver.find_element_by_tag_name('body').send_keys(Keys.END) 
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')
soup.find('table')
rows = soup.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values

data
```
